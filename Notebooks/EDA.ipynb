```python
import pandas as pd 
import numpy as np
import glob
```

```python

all_atp_matches_path = sorted(glob.glob("../Data/tennis_atp/atp_matches_" + "????.csv"))
all_data = None
for match in all_atp_matches_path:
    yearly_data = pd.read_csv(match, parse_dates=[5])
    all_data = pd.concat([all_data, yearly_data])

all_data
```

```python
def readAllFiles():
    allFiles = np.sort(glob.glob("../Data/tennis_atp/atp_matches_" + "????.csv"))[:]
    ranks = pd.DataFrame()
    list_ = list()
    for filen in allFiles:
        df = pd.read_csv(filen,
                         parse_dates=[5])
        list_.append(df)
    ranks = pd.concat(list_)
    return ranks
readAllFiles()
```
